```python
total_sales = df_mart['Sales'].sum()
avg_sales = df_mart['Sales'].mean()
no_of_items_sold = df_mart['Sales'].count()
avg_Profit = df_mart['Profit'].mean()

# Display
print(f"Total sales: ${total_sales:,.0f}")
print(f"Average sales: ${avg_sales:,.0f}")
print(f"Number of items sold: {no_of_items_sold:,.0f}")
print(f"Average Profit: {avg_Profit:.2f}")
```

```python
import matplotlib.pyplot as plt
sales_by_category = df_mart.groupby('Category')['Sales'].sum()

plt.pie(
    sales_by_category, 
    labels=sales_by_category.index, 
    autopct='%.1f%%', 
    startangle=90
)

plt.title('Sales by Category') 
plt.axis('equal')  # Fixed the typo from 'axix' to 'axis'
plt.show()
```

```python
import matplotlib.pyplot as plt
sales_by_category = df_mart.groupby('Category')['Sales'].sum()
sales_by_region = df_mart.groupby('Region')['Sales'].sum()

# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Donut chart 1: Sales by Category
axes[0].pie(
    sales_by_category,
    labels=sales_by_category.index,
    autopct='%.1f%%',
    startangle=90,
    wedgeprops={'width': 0.5}  # creates donut
)
axes[0].set_title('Sales by Category')
axes[0].axis('equal')

# Donut chart 2: Sales by Region
axes[1].pie(
    sales_by_region,
    labels=sales_by_region.index,
    autopct='%.1f%%',
    startangle=90,
    wedgeprops={'width': 0.5}  # creates donut
)
axes[1].set_title('Sales by Region')
axes[1].axis('equal')

# Display
plt.tight_layout()
plt.show()
```

```python
sales_by_type = df_mart.groupby('Region')['Sales'].sum().sort_values(ascending=False)
plt.figure(figsize=(11, 7))
bars = plt.bar(sales_by_type.index, sales_by_type.values, color='gray')
plt.xticks(rotation=90)
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.title('Total Sales by Region')
for bar in bars:
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height(),
        f'{bar.get_height():,.0f}',
        ha='center',
        va='bottom',
        fontsize=12
    )
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Group and reshape the data
grouped = df_mart.groupby(['Category', 'Region'])['Sales'].sum().unstack()

# Filter only 'Regular' and 'Low Fat' categories
grouped = grouped[['North', 'South', 'West', 'Central', 'East']]

# Plot the grouped bar chart
ax = grouped.plot(
    kind='bar',
    figsize=(8, 5),
    title='Outlet Tier by Item Fat Content',
    color=['#FF9999', '#66B3FF','black','gray']  # Optional: custom colors
)

# Axis labels and legend
plt.xlabel('Outlet Location Tier')
plt.ylabel('Total Sales')
plt.legend(title='Item Fat Content')

# Layout adjustment
plt.tight_layout()
plt.show()
```
